```python
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sid = 64
np.random.RandomState(sid)
```

```python
exp_1 = 0
var_1 = 1
alpha = 0.05

norm_distr_1 = norm(exp_1, var_1)
```

```python
N = 100
rnd_gen = np.arange(-6, 8, 0.1)
```

```python
rnd_gen_h = [np.arange(-6, norm_distr_1.ppf(alpha/2), 0.01), np.arange(norm_distr_1.ppf(1 - alpha/2), 8, 0.01)]

sns.lineplot(x=rnd_gen, y=norm_distr_1.pdf(rnd_gen))

plt.vlines(x=norm_distr_1.mean(), ymin =0, ymax = norm_distr_1.pdf(norm_distr_1.mean()), color='r')
plt.vlines(x=norm_distr_1.ppf(alpha/2), ymin =0, ymax = norm_distr_1.pdf(norm_distr_1.ppf(alpha/2)), color='g')
plt.vlines(x=norm_distr_1.ppf(1-alpha/2), ymin =0, ymax = norm_distr_1.pdf(norm_distr_1.ppf(1-alpha/2)), color='g')

plt.fill_between(rnd_gen_h[0], norm_distr_1.pdf(rnd_gen_h[0]), color='g')
plt.fill_between(rnd_gen_h[1], norm_distr_1.pdf(rnd_gen_h[1]), color='g')

plt.show()
```

```python
exp_2 = 3
var_2 = 1

norm_distr_2 = norm(exp_2, var_2)
beta = norm_distr_2.cdf(norm_distr_1.ppf(1 - alpha/2)) * 2
print(beta)
```

```python
sns.lineplot(x=rnd_gen, y=norm_distr_1.pdf(rnd_gen), color='g')
sns.lineplot(x=rnd_gen, y=norm_distr_2.pdf(rnd_gen), color='b')

rnd_gen_b = [np.arange(-6, norm_distr_2.ppf(beta/2), 0.01), np.arange(norm_distr_2.ppf(1 - beta/2), 8, 0.01)]

plt.vlines(x=norm_distr_1.ppf(alpha/2), ymin =0, ymax = norm_distr_1.pdf(norm_distr_1.ppf(alpha/2)), color='g')
plt.vlines(x=norm_distr_1.ppf(1-alpha/2), ymin =0, ymax = norm_distr_1.pdf(norm_distr_1.ppf(1-alpha/2)), color='g')

plt.vlines(x=norm_distr_2.ppf(beta/2), ymin=0, ymax=norm_distr_2.pdf(norm_distr_2.ppf(beta/2)), color='b')
plt.vlines(x=norm_distr_2.ppf(1 - beta/2), ymin=0, ymax=norm_distr_2.pdf(norm_distr_2.ppf(1 - beta/2)), color='b')

plt.fill_between(rnd_gen_h[0], norm_distr_1.pdf(rnd_gen_h[0]), color='g')
plt.fill_between(rnd_gen_h[1], norm_distr_1.pdf(rnd_gen_h[1]), color='g')

plt.fill_between(rnd_gen_b[0], norm_distr_2.pdf(rnd_gen_b[0]), color='b')
plt.fill_between(rnd_gen_b[1], norm_distr_2.pdf(rnd_gen_b[1]), color='b')

plt.show()
```
